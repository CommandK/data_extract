```python
import requests
import json
import pandas as pd


api_key = "62c35cc4e0a9c0.07933082"
symbol = "KO.US"  # Coca Cola's stock symbol
url = f"https://eodhistoricaldata.com/api/fundamentals/{symbol}?api_token={api_key}"

response = requests.get(url)
data = json.loads(response.text)

# print (data)

income_statement_data = data['Financials']['Income_Statement']['quarterly']
balance_sheet_data = data['Financials']['Balance_Sheet']['quarterly']
balance_sheet_data
```

```python
# Open the file in write mode ('w') and dump the JSON data into it
with open('income_statement_data.txt', 'w') as outfile:
    json.dump(income_statement_data, outfile)

# Load the data
with open('income_statement_data.txt') as f:
    data = json.load(f)

```

Create Income Statement Dataframe

```python
# Convert the data into a pandas DataFrame
df = pd.DataFrame.from_dict(data, orient='index')

# Convert the 'date' column to datetime format for easier manipulation later
df['date'] = pd.to_datetime(df['date'])

# Display the first few rows of the DataFrame
df.head()

```
